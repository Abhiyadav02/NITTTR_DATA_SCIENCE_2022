```python
# Creating a Pandas DataFrame

# import pandas as pd
import pandas as pd

# List of String
list_1 = ['data','exploration','using','Pandas','Numpy']

# Create dataframe
df = pd.DataFrame(list_1)

#print dataframe
print(df)
```

```python
# initalised data of list
data = {'Name':['Tom','Nick','Krish','Jack'], 'Age':[21,22,23,24]}

df = pd.DataFrame(data)
print(df)
```

```python
# selecting rows and colums

# initalised data of list
data = {'Name':['Tom','Nick','Krish','Jack'],
        'Age':[21,22,23,24],
       'Address':['Delhi','Kanpur','Jaipur','Noida'],
       'Qualification':['MSC','MA','MCA','Phd']}

# convert the dictionary into DataFrame
df = pd.DataFrame(data)

#select two columns
print(df[['Name','Qualification']])
print() # printing the space between two outputs

# selecting row by integer location method
print(df.iloc[1])  # retrieve row and columns by position
```

### Working with Missing data

#### MIssing data can occur when no information is provided for one or more items or for a whole unit. Missing Data is a very big problem in real life scenario. Missing Data 

```python
import numpy as np
import pandas as pd

# Dictionary of lists
dict_1 = {'First Score':[100,90,np.nan,96],
         'Second Score':[30,45,56, np.nan],
         'Thirs Score':[np.nan, 40, 80, 98]}

# creating a dataframe from list
df = pd.DataFrame(dict_1)

# using isnull() function to check the null values are present or not
df.isnull()
```

### Filling missing values using fillna(), replace() and interpolate()

```python
import numpy as np
import pandas as pd

# Dictionary of lists
dict_1 = {'First Score':[100,90,np.nan,96],
         'Second Score':[30,45,56, np.nan],
         'Thirs Score':[np.nan, 40, 80, 98]}

# creating a dataframe from list
df = pd.DataFrame(dict_1)

# using isnull() function to check the null values are present or not
df.isnull()

# filling missing values using fillna()
df.fillna(0)
```

## dropna() function

```python
import numpy as np
import pandas as pd

# Dictionary of lists
dict_1 = {'First Score':[100,90,np.nan,96],
         'Second Score':[30,45,56, np.nan],
         'Thirs Score':[np.nan, 40, 80, 98]}

# creating a dataframe from list
df = pd.DataFrame(dict_1)
df

# using dropna() function drops the all colums or orws which contains null value
df.dropna()
```
